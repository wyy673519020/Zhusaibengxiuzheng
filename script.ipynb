```python
elev1= 0
azim1= 0
elev2= 0
azim2= 90
elev3=90
azim3= 0
elev4=22.0
azim4=225.5
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,16),dpi=80)
rect1 = [0.1, 0.5, 0.4, 0.4]
rect2 = [0.5, 0.5, 0.4, 0.4]
rect3 = [0.1, 0.1, 0.4, 0.4]
rect4 = [0.5, 0.1, 0.4, 0.4]
data1=np.genfromtxt('data/source_std1.csv',delimiter=',')
data2=np.genfromtxt('data/source_std2.csv',delimiter=',')
xline1= data1[0:646,1]
yline1= data1[0:646,2]
zline1= data1[0:646,3]
xline2=data2[0:42,1]
yline2=data2[0:42,2]
zline2=data2[0:42,3]
ax1 = plt.axes(rect1,projection='3d')
ax2 = plt.axes(rect2,projection='3d')
ax3 = plt.axes(rect3,projection='3d')
ax4 = plt.axes(rect4,projection='3d')
ax1.view_init(elev1, azim1)
ax1.scatter(xline1,yline1,zline1,c='g', marker='o',s=20)
ax1.scatter(xline2,yline2,zline2,c='r', marker='p',s=200)
#plt.grid(alpha='1.9')  #配置网格线
ax1.set_xlabel('Np')
ax1.set_ylabel('deltaP')
ax1.set_zlabel('W')
ax2.view_init(elev2, azim2)
ax2.scatter(xline1,yline1,zline1,c='g', marker='o',s=20)
ax2.scatter(xline2,yline2,zline2,c='r', marker='p',s=200)
#plt.grid(alpha='1.9')  #配置网格线
ax2.set_xlabel('Np')
ax2.set_ylabel('deltaP')
ax2.set_zlabel('W')
ax3.view_init(elev3, azim3)
ax3.scatter(xline1,yline1,zline1,c='g', marker='o',s=20)
ax3.scatter(xline2,yline2,zline2,c='r', marker='p',s=200)
#plt.grid(alpha='1.9')  #配置网格线
ax3.set_xlabel('Np')
ax3.set_ylabel('deltaP')
ax3.set_zlabel('W')
ax4.view_init(elev4, azim4)
ax4.scatter(xline1,yline1,zline1,c='g', marker='o',s=20)
ax4.scatter(xline2,yline2,zline2,c='r', marker='p',s=200)
#plt.grid(alpha='1.9')  #配置网格线
ax4.set_xlabel('Np')
ax4.set_ylabel('deltaP')
ax4.set_zlabel('W')
plt.suptitle("Pout=20.4082",fontsize=20, color='m')
plt.show()
```
